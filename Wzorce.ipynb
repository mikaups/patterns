### Singleton

```python
class Singleton:
    instance = None

    def __new__(cls, *args, **kwargs):
        if not isinstance(cls.instance, cls):
            cls.instance = object.__new__(cls)
        return cls.instance
```

### Prototyp

```python
class Prototyp:
    def __init__(self):
        self.parent = None

    def set_parent(self, parent):
        self.parent = parent
        
class PrototypsComponent:
    def __init__(self, some_int, some_list, some_ref):
        self.some_int = some_int
        self.some_list = some_list
        self.some_ref = some_ref

    def __copy__(self):
        some_list = copy.copy(self.some_list)
        some_ref = copy.copy(self.some_ref)

        new = self.__class__(
            self.some_int, some_list, some_ref
        )
        new.__dict__.update(self.__dict__)

        return new

    def __deepcopy__(self, memo={}):
        some_list = copy.deepcopy(self.some_list, memo)
        some_ref = copy.deepcopy(self.some_ref, memo)

        new = self.__class__(
            self.some_int, some_list, some_ref
        )
        new.__dict__ = copy.deepcopy(self.__dict__, memo)

        return new


if __name__ == "__main__":

    mylist = [1, {1, 2, 3}, [1, 2, 3]]
    circular_ref = Prototyp()
    component = PrototypsComponent(23, mylist, circular_ref)
    circular_ref.set_parent(component)

    shallow_copy = copy.copy(component)
    shallow_copy.some_list.append("kolejny obiekt")
    if component.some_list[-1] == "kolejny obiekt":
        print(
            "Dodawanie do listy shallow_copy"
            "dodaje element do `component` "
        )
    else:
        print(
            "Dodawanie do listy shallow_copy"
            "nie dodaje elementu do `component` "
        )

    component.some_list[1].add(4)
    if 4 in shallow_copy.some_list[1]:
        print(
            "Zmienianie obiektu w liscie `component`"
            "zmienia go w liscie `shallow_copied_component` "
        )
    else:
        print(
            "Zmienianie obiektu w liscie `component`"
            "nie zmienia go w liscie `shallow_copied_component` "
        )

    deep_copy = copy.deepcopy(component)

    deep_copy.some_list.append("kolejny obiekt")
    if component.some_list[-1] == "kolejny obiekt":
        print(
            "Dodawanie do listy deep_copy"
            "dodaje element do `component` "
        )
    else:
        print(
            "Dodawanie do listy deep_copy"
            "dodaje element do `component` "
        )

    component.some_list[1].add(10)
    if 10 in deep_copy.some_list[1]:
        print(
            "Zmienianie obiektu w liscie `component`"
            "zmienia go w liscie `deep_copy` "
        )
    else:
        print(
            "Zmienianie obiektu w liscie `component`"
            "nie zmienia go w liscie `deep_copy` "
        )

    print(
        f"id(deep_copy.some_ref.parent): "
        f"{id(deep_copy.some_ref.parent)}"
    )
    print(
        f"id(deep_copy.some_ref.parent.some_ref.parent): "
        f"{id(deep_copy.some_ref.parent.some_ref.parent)}"
    )
    print(
        "Deep copied obiekty zawierają to samo odniesienie, nie są wielokrotnie klonowane"
    )
```

### Fasada

```python
class Fasada:
    def __init__(self, plyta: Plyta, sektorA: SektorA, sektorB: SektorB, sektorC: SektorC) -> None:
        self.plyta = plyta or Plyta()
        self.sektorA = sektorA or SektorA()
        self.sektorB = sektorB or SektorB()
        self.sektorC = sektorC or SektorC()

    def operation(self) -> bool:
        results = []
        results.append("Sprawdzanie czy sa jeszcze wolne miejsc na koncert:")
        results.append(self.plyta.operation1())
        results.append(self.sektorA.operation1())
        results.append(self.sektorB.operation1())
        results.append(self.sektorC.operation1())
        results.append("Dostepne miejsca:")
        results.append(self.plyta.operation_n())
        results.append(self.sektorA.operation_z())
        results.append(self.sektorB.operation_x())
        results.append(self.sektorC.operation_y())

        return "\n".join(results)


class Plyta:
    def operation1(self) -> str:
        if True:
            return "Tak"
    def operation_n(self) -> str:
        return "Miejsce na plycie jest dostepne"


class SektorA:
    def operation1(self) -> str:
        if True:
            return "Tak"
    def operation_z(self) -> str:
        return "Miejsce w sektorze A jest dostepne"

class SektorB:
    def operation1(self) -> str:
        if True:
            return "Tak"
    def operation_x(self) -> str:
        return "Miejsce w sektorze B jest dostepne"
    
class SektorC:
    def operation1(self) -> str:
        if True:
            return "Tak"
    def operation_y(self) -> str:
        return "Miejsce w sektorze C jest dostepne"
    
def client_code(facade: Facade) -> None:
    print(facade.operation(), end="")


if __name__ == "__main__":

    plyta = Plyta()
    sektorA = SektorA()
    sektorB = SektorB()
    sektorC = SektorC()
    fasada = Fasada(plyta, sektorA, sektorB, sektorC)
    client_code(fasada)
```

### Pełnomocnik

### Kompozyt

```python
class kompozyt(ABC):
    @property
    def parent(self) -> kompozyt:
        return self._parent

    @parent.setter
    def parent(self, parent: kompozyt):
        self._parent = parent

    def dodaj(self, kompozyt: kompozyt) -> None:
        pass

    def usun(self, kompozyt: kompozyt) -> None:
        pass

    def czy_kompozyt(self) -> bool:
        return False

    @abstractmethod
    def op(self) -> str:
        pass


class Lisc(kompozyt):
    def op(self) -> str:
        return "lisc"


class komp(kompozyt):
    def __init__(self) -> None:
        self._children: List[kompozyt] = []
            
    def dodaj(self, kompozyt: kompozyt) -> None:
        self._children.append(kompozyt)
        kompozyt.parent = self

    def usun(self, kompozyt: kompozyt) -> None:
        self._children.usun(kompozyt)
        kompozyt.parent = None

    def is_composite(self) -> bool:
        return True

    def op(self) -> str:
        res = []
        for child in self._children:
            res.append(child.op())
        return f"Kolejny lisc {' + '.join(res)}"


def client_code(kompozyt: kompozyt) -> None:
    print(f"Poczatek: {kompozyt.op()}", end="")


def client_code2(kompozyt1: kompozyt, kompozyt2: kompozyt) -> None:
    if kompozyt1.czy_kompozyt():
        kompozyt1.dodaj(kompozyt2)

    print(f"Mamy drzewo:  {kompozyt1.op()}", end="")


if __name__ == "__main__":
    lisc = Lisc()
    print("Na poczatku byl jeden listek:")
    client_code(lisc)
    print("\n")

    drzewo = komp()

    galaz = komp()
    galaz.dodaj(Lisc())
    galaz.dodaj(Lisc())

    galaz2 = komp()
    galaz2.dodaj(Lisc())

    drzewo.dodaj(galaz)
    drzewo.dodaj(galaz2)

    print("Potem bylo wiecej listkow:")
    client_code(drzewo)
    print("\n")

    print("Az w koncu...")
    client_code2(drzewo, lisc)
```

### Iterator

```python
class Iterator(Iterator):

    _miejsce: int = None
    _wSpak: bool = False

    def __init__(self, kolekcja: Kolekcja, wSpak: bool = False) -> None:
        self._kolekcja = kolekcja
        self._wSpak = wSpak
        self._miejsce = -1 if wSpak else 0

    def __next__(self):
        try:
            v = self._kolekcja[self._miejsce]
            self._miejsce += -1 if self._wSpak else 1
        except IndexError:
            raise StopIteration()

        return v


class Kolekcja(Iterable):

    def __init__(self, kolekcja: List[Any] = []) -> None:
        self._kolekcja = kolekcja

    def __iter__(self) -> Iterator:
        return Iterator(self._kolekcja)

    def naOdwot(self) -> Iterator:
        return Iterator(self._kolekcja, True)

    def dodaj(self, item: Any):
        self._kolekcja.append(item)


if __name__ == "__main__":

    kolekcja = Kolekcja()
    kolekcja.dodaj("Pierwszy wyraz")
    kolekcja.dodaj("Drugi wyraz")
    kolekcja.dodaj("Trzeci wyraz")

    print("Normalny porządek:")
    print("\n".join(kolekcja))
    print("")

    print("A tu wszystko na odwrót:")
    print("\n".join(kolekcja.naOdwot()), end="")
```

### Pamiątka

```python
class Klient():
    _state = None

    def __init__(self, stan: str) -> None:
        self._stan = stan
        print(f"Klient: Zglaszam problem o kodzie: {self._stan}")

    def do_something(self) -> None:
        self._stan = self._generate_random_string(30)
        print(f"Klient: Otrzymany kod: {self._stan}")

    def _generate_random_string(self, length: int = 10) -> None:
        return "".join(sample(ascii_letters, length))

    def save(self) -> Memento:
        return ConcreteMemento(self._stan)

    def restore(self, memento: Memento) -> None:
        self._stan = memento.get_stan()
        print(f"Klient: Otrzymany kod zgloszenia: {self._stan}")


class Memento(ABC):
    @abstractmethod
    def get_nazwa(self) -> str:
        pass

    @abstractmethod
    def get_data(self) -> str:
        pass


class ConcreteMemento(Memento):
    def __init__(self, stan: str) -> None:
        self._stan = stan
        self._data = str(datetime.now())[:19]

    def get_stan(self) -> str:
        return self._stan

    def get_nazwa(self) -> str:
        return f"{self._data} / ({self._stan[0:9]}...)"

    def get_data(self) -> str:
        return self._data


class System():
    def __init__(self, system: System) -> None:
        self._mementos = []
        self._system = system

    def backup(self) -> None:
        print("\nSystem: Na skrzynke mailowa zostala wyslana wiadomosc z indywidualnymi kodami, prosze je wpisac")
        self._mementos.append(self._system.save())

    def undo(self) -> None:
        if not len(self._mementos):
            return

        memento = self._mementos.pop()
        print(f"System: Ostatnie zgloszenie: {memento.get_nazwa()}")
        try:
            self._system.restore(memento)
        except Exception:
            self.undo()

    def show_history(self) -> None:
        print("System: Stan zgloszen:")
        for memento in self._mementos:
            print(memento.get_nazwa())


if __name__ == "__main__":
    klient = Klient("A3")
    system = System(klient)

    system.backup()
    klient.do_something()

    system.backup()
    klient.do_something()

    print()
    system.show_history()

    print()
    system.undo()

    print()
    system.undo()
```

### Łańcuch zobowiązań

```python
class Lancuch_zobowiazan(ABC):
    @abstractmethod
    def nastepny(self, lancuch_zobowiazan: Lancuch_zobowiazan) -> Lancuch_zobowiazan:
        pass

    @abstractmethod
    def jedzenie(self, request) -> Optional[str]:
        pass


class Posilek(Lancuch_zobowiazan):
    dalej: Lancuch_zobowiazan = None

    def nastepny(self, lancuch_zobowiazan: Lancuch_zobowiazan) -> Lancuch_zobowiazan:
        self.dalej = lancuch_zobowiazan
        return lancuch_zobowiazan

    @abstractmethod
    def jedzenie(self, request: Any) -> str:
        if self.dalej:
            return self.dalej.jedzenie(request)
        return None

class Malpka(Posilek):
    def jedzenie(self, request: Any) -> str:
        if request == "Banana":
            return f"Malpka: ja chce {request}"
        else:
            return super().jedzenie(request)

class Wiewiorka(Posilek):
    def jedzenie(self, request: Any) -> str:
        if request == "Orzechy":
            return f"Wiewiorka: ja chce {request}"
        else:
            return super().jedzenie(request)

class Pies(Posilek):
    def jedzenie(self, request: Any) -> str:
        if request == "Kosc":
            return f"Pies: ja chce {request}"
        else:
            return super().jedzenie(request)

def client_code(lancuch_zobowiazan: Lancuch_zobowiazan) -> None:
    for food in ["Orzechy", "Banana", "Warzywa", "Kosc"]:
        print(f"\nOpiekun: Kto chce {food}?")
        result = lancuch_zobowiazan.jedzenie(food)
        if result:
            print(f"  {result}", end="")
        else:
            print(f"  {food} jakos nie ciesza sie popularnoscia", end="")

if __name__ == "__main__":
    malpka = Malpka()
    wiewiorka = Wiewiorka()
    pies = Pies()

    malpka.nastepny(wiewiorka).nastepny(pies)
    
    print("Kolejnosc: Małpka > Wiewiorka > Pies")
    client_code(malpka)
    print("\n")

    print("Fragment lancucha: Wiewiorka > Pies")
    client_code(wiewiorka)
```

### Obserwator

```python
class Gielda(ABC):

    @abstractmethod
    def dodanie(self, inwestor: Inwestor) -> None:
        pass

    @abstractmethod
    def usuniecie(self, inwestor: Inwestor) -> None:
        pass

    @abstractmethod
    def powiadomienie(self) -> None:
        pass


class Obsluga_inwestorow(Gielda):

    _state: int = None
    _observers: List[Observer] = []

    def dodanie(self, inwestor: Inwestor) -> None:
        print("Dodawanie nowego inwestora.")
        self._observers.append(inwestor)

    def usuniecie(self, inwestor: Inwestor) -> None:
        self._observers.remove(inwestor)

    def powiadomienie(self) -> None:

        print("Powiadomienie inwestorow...")
        for observer in self._observers:
            observer.update(self)

    def pieniadz_pracuje(self) -> None:

        print("\nUaktualnianie stanu gieldy.")
        self._state = randrange(0, 10)

        print(f"Aktualna wartosc akcji: {self._state}")
        self.powiadomienie()


class Inwestor(ABC):

    @abstractmethod
    def update(self, gielda: Gielda) -> None:
        pass

class InwestorA(Observer):
    def update(self, gielda: Gielda) -> None:
        if gielda._state < 3:
            print("InwestorA: Kupuje wiecej!")


class InwestorB(Observer):
    def update(self, gielda: Gielda) -> None:
        if gielda._state == 0 or gielda._state >= 2:
            print("InwestorB: Rezygnuje")


if __name__ == "__main__":
    oInwestorow = Obsluga_inwestorow()

    observer_a = InwestorA()
    oInwestorow.dodanie(observer_a)

    observer_b = InwestorB()
    oInwestorow.dodanie(observer_b)

    oInwestorow.pieniadz_pracuje()
    oInwestorow.pieniadz_pracuje()

    oInwestorow.usuniecie(observer_b)

    oInwestorow.pieniadz_pracuje()
```

### Polecenie

```python
class Polecenie(ABC):
    @abstractmethod
    def execute(self) -> None:
        pass

class Dok1(Polecenie):
    def __init__(self, payload: str) -> None:
        self._payload = payload

    def execute(self) -> None:
        print(f"Ja musze tylko wydrukować "
              f"{self._payload}")

class DokN(Polecenie):
    def __init__(self, receiver: Receiver, a: str, b: str) -> None:
        self._receiver = receiver
        self._a = a
        self._b = b

    def execute(self) -> None:
        print("Ja!", end="")
        self._receiver.kol(self._a)
        self._receiver.kolN(self._b)


class OutP:
    def kol(self, a: str) -> None:
        print(f"\nRKolejka wydruku {a}.", end="")

    def kolN(self, b: str) -> None:
        print(f"\nKolejka wydruku {b}.", end="")


class Osoba:
    _start = None
    _koniec = None
    def set_start(self, polecenie: Polecenie):
        self._start = polecenie

    def set_koniec(self, polecenie: Polecenie):
        self._koniec = polecenie

    def kolI(self) -> None:
        print("Ktoś korzysta teraz z drukarki?")
        if isinstance(self._start, Polecenie):
            self._start.execute()
        print("Ktos jeszcze?")
        if isinstance(self._koniec, Polecenie):
            self._koniec.execute()


if __name__ == "__main__":
    osoba = Osoba()
    osoba.set_start(Dok1("Dokument1"))
    OutP = OutP()
    osoba.set_koniec(DokN(
        OutP, "Dokument2", "Dokument3"))

    osoba.kolI()
```
